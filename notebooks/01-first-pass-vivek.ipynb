```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Avoid division by zero for distance
df = df[df['trip_miles'] > 0]

# Compute base fare per mile
df['fare_per_mile'] = df['base_passenger_fare'] / df['trip_miles']

# Remove outliers using IQR method
Q1 = df['fare_per_mile'].quantile(0.25)
Q3 = df['fare_per_mile'].quantile(0.75)
IQR = Q3 - Q1

# Define valid range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df = df[(df['fare_per_mile'] >= lower_bound) & (df['fare_per_mile'] <= upper_bound)]

# Define price bins for fare per mile
bins = np.linspace(df['fare_per_mile'].min(), df['fare_per_mile'].max(), num=15)  # 15 bins
df['fare_bin'] = pd.cut(df['fare_per_mile'], bins)

# Aggregate Demand: Count of trips per fare per mile bin
demand = df.groupby('fare_bin').size().reset_index(name='trip_count')

# Aggregate Supply: Use 1 / wait_time
df['inverse_wait_time'] = 1 / df['wait_time'].replace(0, np.nan)  # Avoid division by zero
supply = df.groupby('fare_bin')['inverse_wait_time'].mean().reset_index(name='supply_metric')

# Merge supply and demand
supply_demand = demand.merge(supply, on="fare_bin", how="left")

# Convert categorical bins to midpoints for plotting
supply_demand['fare_mid'] = supply_demand['fare_bin'].apply(lambda x: x.mid)

# Min-Max Scaling (0 to 1) for demand and supply
supply_demand['trip_count_scaled'] = (supply_demand['trip_count'] - supply_demand['trip_count'].min()) / \
                                     (supply_demand['trip_count'].max() - supply_demand['trip_count'].min())

supply_demand['supply_scaled'] = (supply_demand['supply_metric'] - supply_demand['supply_metric'].min()) / \
                                 (supply_demand['supply_metric'].max() - supply_demand['supply_metric'].min())

# Plot the supply and demand curves (Scaled)
plt.figure(figsize=(8, 5))
plt.plot(supply_demand['fare_mid'], supply_demand['trip_count_scaled'], marker='o', label='Demand (Scaled)', color='blue')
plt.plot(supply_demand['fare_mid'], supply_demand['supply_scaled'], marker='s', label='Supply (Scaled 1/Wait Time)', color='red')

plt.xlabel("Base Fare per Mile ($)")  # Raw price on X-axis
plt.ylabel("Scaled Quantity (0-1)")  # Scaled supply & demand on Y-axis
plt.title("NYC Taxi Supply-Demand Curve (Scaled Supply & Demand, Raw Price)")
plt.legend()
plt.grid()
plt.show()
```
